# Import Required Libraries
Import the necessary Python libraries for handling credential data.

```python
import bcrypt
from bcrypt import hashpw, gensalt, checkpw
```

# Define Current User Credentials

# Update User Name and Password
Change the user name to `angelo` and compute the password hash for `angelo123`.

```python
new_password_hash = hashpw(new_password.encode('utf-8'), gensalt())
new_password_hash_str = new_password_hash.decode('utf-8')
updated_credentials = {
    'username': new_username,
    'password_hash': new_password_hash_str
}
updated_credentials
```
